```python
class number:
    
    def __init__(self,number,number_of_subsections,min_random_number_desired):
        self.number = number
        self.number_of_subsections =number_of_subsections
        self.min_random_number_desired = min_random_number_desired
    #def accept(self,number, number_of_subsections, min_random_number_desired):
   
 
        #ob = number(number, number_of_subsections, min_random_number_desired)
        #ls.append(ob)
    

    def split_given_number_into_n_random_numbers(self):
        a=[]
        cumulative_sum_of_random_numbers = 0
        current_subsection = 1
        max_random_number = int(self.number/self.number_of_subsections)
        if self.min_random_number_desired > max_random_number:
            print("ERROR: Cannot have min number as {} and split {} in {} subsections".format(self.min_random_number_desired,
                                                                                              self.number, self.number_of_subsections))
            return False

        while (True):
            random_number = random.randint(self.min_random_number_desired, max_random_number)
            a.append(random_number)
            cumulative_sum_of_random_numbers += random_number
            #print("Cumulative sum {}".format(sum_of_num))
            self.number -= random_number
            current_subsection += 1
            if current_subsection == self.number_of_subsections:
                random_number = self.number
                a.append(random_number)
                cumulative_sum_of_random_numbers += random_number
                break
        #print(a)

        #print("Final cumulative sum of random numbers = {}".format(cumulative_sum_of_random_numbers))
        return a



#obj = number(58,5,0)
#obj.accept(58,7,0)
number(58,5,0).split_given_number_into_n_random_numbers()
```
